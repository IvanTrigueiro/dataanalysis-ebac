# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
```

```
data.head(20)
```

```
print(data.shape[0])
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

### resposta da questão 1.1

```
# Checar se existem dados faltando em mais de uma coluna de uma linha
missing_data_rows = data.isnull().sum(axis=1) > 1

# Mostrar as linhas que têm dados faltantes em mais de uma coluna
rows_with_multiple_missing_values = data[missing_data_rows]
print(f"{rows_with_multiple_missing_values} \n\n {missing_data_rows.sum()}")
```

```
# Como só tem duas linhas com dados faltantes em mais de uma coluna, não vejo problema em remover essas linhas.
# Identificar as linhas que precisam ser removidas
lines_to_remove = [3, 339]
# Remover as linhas específicas
data_cleaned = data.drop(lines_to_remove)
# Resetar o índice
data_cleaned.reset_index(drop=True, inplace=True)
data_cleaned
```

#### Agora vamos tratar os dados faltando na coluna sex.

```
data_cleaned = data_cleaned.dropna(subset=["sex"])
data_cleaned.reset_index(drop=True, inplace=True)
data_cleaned
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

#### Resposta da questão 1.2

```
df = data_cleaned
```

```
df_bill_length_mean = df["bill_length_mm"].mean()
df_bill_depth_mean = df["bill_depth_mm"].mean()
df_flipper_length_mean = df["flipper_length_mm"].mean()
df_body_mass_g_mean = df["body_mass_g"].mean()

df_bill_length_std =  df["bill_length_mm"].std()
df_bill_depth_std = df["bill_depth_mm"].std()
df_flipper_length_std = df["flipper_length_mm"].std()
df_body_mass_g_std = df["body_mass_g"].std()
```

```
df['bill_length_mm_std'] = df['bill_length_mm'].apply(lambda nota: (nota - df_bill_length_mean) / df_bill_length_std)
df['bill_depth_mm_std'] = df['bill_depth_mm'].apply(lambda nota: (nota - df_bill_depth_mean) / df_bill_depth_std)
df['flipper_length_mm_std'] = df['flipper_length_mm'].apply(lambda nota: (nota - df_flipper_length_mean) / df_flipper_length_std)
df['body_mass_g_std'] = df['body_mass_g'].apply(lambda nota: (nota - df_body_mass_g_mean) / df_body_mass_g_std)
```

```
df
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

#### Resposta da questão 1.3

```
df['island'].drop_duplicates()
```

```
df['species'].drop_duplicates()
```

```
species_mapper = {
    "Adelie": 0,
    "Chinstrap": 1,
    "Gentoo": 2
}

islands_mapper = {
    "Torgersen": 0,
    "Biscoe": 1,
    "Dream": 2
}

sex_mapper = {
    "Male": 0,
    "Female": 1
}
```

```
df["species_ord"] = df["species"].apply(lambda x: species_mapper[x])
df["island_ord"] = df["island"].apply(lambda x: islands_mapper[x])
df["sex_nom"] = df["sex"].apply(lambda x: sex_mapper[x])
```

```
df
```

### **1.4. Limpeza**

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".

#### Resposta da questão 1.4

```
df.drop(columns=["species",	"island",	"bill_length_mm",	"bill_depth_mm",	"flipper_length_mm", "body_mass_g", "sex"], inplace=True)
```
